```python
def fetch_fred_data_to_df(series_dict):
    """
    Fetches data for given FRED series IDs and returns a DataFrame with each series as a column.

    Parameters:
    - series_dict: A dictionary where keys are FRED series IDs and values are series descriptions.

    Returns:
    - A pandas DataFrame where each column represents a series, named by its description.
    """
    # Initialize an empty DataFrame
    df = pd.DataFrame()

    for series_id, description in series_dict.items():
        try:
            # Fetch the series data
            series_data = fred.get_series(series_id)

            # If the DataFrame is empty, initialize it with the fetched data
            if df.empty:
                df = pd.DataFrame(series_data, columns=[description])
            else:
                # If the DataFrame already exists, add this series as a new column
                df[description] = series_data
        except Exception as e:
            print(f"Failed to fetch data for series ID {series_id} ({description}): {e}")

    return df
```
